#  FEATURE CREATION
1. Create Tenure group.
2. Create Monthly charges category.
3. Create Total spending category.
4. Create Contract type category.

```python
bins = [0, 12, 24, 48, 60, 72]
labels = ['0-1 Year', '1-2 Years', '2-4 Years', '4-5 Years', '5-6 Years']

df['Tenure_Group'] = pd.cut(df['tenure'], bins=bins, labels=labels)
```

```python
bins = [0, 35, 70, 120]
labels = ['Low', 'Medium', 'High']

df['MonthlyCharges_Category'] = pd.cut(df['MonthlyCharges'], bins=bins, labels=labels)
```

```python
bins = [0, 2000, 5000, 9000]
labels = ['Low Spender', 'Medium Spender', 'High Spender']

df['TotalSpending_Category'] = pd.cut(df['TotalCharges'], bins=bins, labels=labels)
```

# FEATURE SCALING
Apply suitable scaling techniques to Monthly Charges, Total Charges, and Tenure.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['MonthlyCharges', 'TotalCharges', 'tenure']] = scaler.fit_transform(
    df[['MonthlyCharges', 'TotalCharges', 'tenure']]
)
```

# EXPLORATORY DATA ANALYSIS
1. Identify which contract type has the highest churn.
2. Check whether higher monthly charge increases churn.
3. Check whether longer tenure reduces churn.
4. Identify which payment method has the highest churn rate.

```python
df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()
```

##  Retention Strategies

Based on the EDA, the following strategies can help reduce customer churn:

### 1. Target High-Risk Customers
- Focus on customers with **month-to-month contracts**, **high monthly charges**, and **low tenure**.
- Identify these customers for personalized retention campaigns.

### 2. Offer Incentives to Switch to Long-Term Contracts
- Provide **discounted yearly plans**.
- Implement **loyalty programs** to encourage longer commitments.

### 3. Promote Add-On Services
- Encourage adoption of **TechSupport, DeviceProtection, OnlineSecurity**, etc.
- Customers with **more services** are less likely to churn.

### 4. Payment Method Adjustments
- Encourage **automatic payments** instead of **electronic check**.
- Reduce failed payments and payment-related issues.

### 5. Customer Engagement
- Send **proactive emails** for engagement and updates.
- Provide **rewards for tenure** or consistent usage.
- Offer **early support for complaints** to improve customer satisfaction.
